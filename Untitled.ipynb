```python
def get_selected_rows(df, start_date,end_date):
    df['temp'] = pd.to_datetime(df['Created'])
    df = df[(df['temp'] >= start_date) & (df['temp'] <= end_date)]
    df.pop('temp')
    return df
```

```python
start_date = "2022-06-14"
end_date = "2022-06-17"
get_selected_rows(df, start_date,end_date)
```
